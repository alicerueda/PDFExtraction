```python
# a function to extract values from a pdf file given the key_mapping dictionary
import PyPDF2
def extract_values_from_pdf(pdf_path, key_mapping):
    values = {}
    mapped_values = {}
    
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]#pdf_reader.getPage(page_num)
            
            # Extract annotations (form fields) from the page
            annotations = page.get('/Annots')
            
            if annotations:
                for annotation in annotations:
                    if annotation:
                        # Check if the annotation is a checkbox
                        if annotation.get_object()['/Subtype'] == '/Widget' and annotation.get_object()['/Parent'].get_object()['/FT'] == '/Btn':
                            field_name = annotation.get_object()['/Parent'].get_object()['/T']
                            #print(field_name)
                            if '/V' in annotation.get_object()['/Parent'].get_object():
                                # Check if the checkbox is checked
                                is_checked = annotation.get_object()['/Parent'].get_object()['/V'][:-5] == '/Yes'
                            
                                # Store the checkbox value in the dictionary
                                values[field_name] = is_checked
                                if field_name[:-4] in key_mapping:
                                    mapped_values[key_mapping[field_name[:-4]][:-2]] = key_mapping[field_name[:-4]][-2:]#is_checked
                            
                        # Check if the annotation is a text field
                        elif annotation.get_object()['/Subtype'] == '/Widget' and annotation.get_object()['/Parent'].get_object()['/FT'] == '/Tx':
                            field_name = annotation.get_object()['/Parent'].get_object()['/T']
                            #print(field_name)
                            # Extract the text field value
                            if '/V' in annotation.get_object()['/Parent'].get_object():
                                field_value = annotation.get_object()['/Parent'].get_object()['/V']
                            
                                # Store the text field value in the dictionary
                                values[field_name] = field_value
                                if field_name[:-4] in key_mapping:
                                    mapped_values[key_mapping[field_name[:-4]]] = field_value
    
    return mapped_values
```

```python
# a function to extract values from a pdf file given the key_mapping dictionary
def CRF_extract_values_from_pdf(pdf_path, key_mapping):
    values = {}
    mapped_values = {}
    
    with open(pdf_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]#pdf_reader.getPage(page_num)
            
            # Extract annotations (form fields) from the page
            #annotations = page.get('/Annots')
            annotations = page.get('/Annots').get_object()
            if annotations:
                for annotation in annotations:
                    if annotation:
                        # Check if the annotation is a checkbox
                        #print(annotation.get_object()['/Parent'].get_object()['/Kids'][0].get_object())
                        if annotation.get_object()['/Parent'].get_object()['/Kids'][0].get_object()['/Subtype'] == '/Widget' and annotation.get_object()['/Parent'].get_object()['/Kids'][0].get_object()['/FT'] == '/Btn':
                            field_name = annotation.get_object()['/Parent'].get_object()['/T']
                            #print(field_name)
                            #if '/V' in annotation.get_object()['/Parent'].get_object():
                            if '/V' in annotation.get_object()['/Parent'].get_object()['/Kids'][0].get_object():
                                # Check if the checkbox is checked
                                is_checked = annotation.get_object()['/Parent'].get_object()['/Kids'][0].get_object() == '/Yes'
                            
                                # Store the checkbox value in the dictionary
                                values[field_name] = is_checked
                                if field_name[:-4] in key_mapping:
                                    mapped_values[key_mapping[field_name[:-4]].split("_")[0]] = key_mapping[field_name[:-4]].split("_")[1]
                            
                        # Check if the annotation is a text field
                        elif annotation.get_object()['/Parent'].get_object()['/Kids'][0].get_object()['/Subtype'] == '/Widget' and annotation.get_object()['/Parent'].get_object()['/Kids'][0].get_object()['/FT'] == '/Tx':
                            field_name = annotation.get_object()['/Parent'].get_object()['/T']
                            #print(field_name)
                            # Extract the text field value
                            #if '/V' in annotation.get_object()['/Parent'].get_object():
                            if '/V' in annotation.get_object()['/Parent'].get_object()['/Kids'][0].get_object():
                                #field_value = annotation.get_object()['/Parent'].get_object()['/V']
                                field_value = annotation.get_object()['/Parent'].get_object()['/Kids'][0].get_object()['/V']
                                # Store the text field value in the dictionary
                                values[field_name] = field_value
                                if field_name[:-4] in key_mapping:
                                    mapped_values[key_mapping[field_name[:-4]]] = field_value
    
    return mapped_values
```
